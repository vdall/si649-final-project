```
import altair as alt
import pandas as pd
from vega_datasets import data
df = pd.read_csv("https://raw.githubusercontent.com/valxwz/data-viz/master/gp/MoMA_artworks_cleaned_important_columns.csv")
df.sample(5)
```

```
gender_d=df[df.Gender!='Non-Binary']
gender_df=gender_d.groupby(['Gender','Classification','DateAcquired']).count().sort_values(['Gender'],ascending=False).reset_index()
gender_year_df=gender_df[gender_df['DateAcquired'].between(1940.0,2019.0)]
gender_year_df
```

```
pink_blue = alt.Scale(domain=('Male', 'Female'),
                      range=["#6DCCDA", "#FE9E4A"])
slider = alt.binding_range(min=1940.0, max=2019.0, step=1,name="Year")
select_year = alt.selection_single(name="DateAcquired", fields=['DateAcquired'],
                                   bind=slider, init={'DateAcquired':2019.0})
chart1=alt.Chart(gender_year_df,width=50).mark_bar().encode(
    x=alt.X('Gender:O', axis=alt.Axis(title=None,grid=False,ticks=False,labels=False)),
    y=alt.Y('Title:Q', axis=alt.Axis(title=None,grid=False),scale=alt.Scale(domain=(1,1100))),
    tooltip="Title:N",
    color=alt.Color('Gender:N',scale=pink_blue),
    column=alt.Column('Classification:N',title=None,header=alt.Header(labelAngle=45,labelFontSize=12,orient='bottom',labelPadding=25,labelAlign='left'),sort=["Photograph","Print","Drawing","Painting","Sculpture"])
).add_selection(
    select_year
).transform_filter(
    select_year
).transform_filter(
    alt.FieldOneOfPredicate(field='Classification',oneOf=["Photograph","Print","Drawing","Painting","Sculpture"])
)
chart1
```

```
alt.Chart(gender_year_df,width=50, title="Classifications of Artworks Acquired by Gender in Total").mark_bar().encode(
    x=alt.X('Gender:O',
            axis=alt.Axis(title=None,grid=False,ticks=False,labels=False)),
    y=alt.Y('Title:Q',stack=None,axis=alt.Axis(title=None,grid=False,ticks=False,labels=True)),
    tooltip="Title:N",
    color=alt.Color('Gender:N',scale=pink_blue),
    column=alt.Column('Classification:N',title=None,header=alt.Header(labelAngle=45,labelFontSize=12,orient='bottom',labelPadding=30,labelAlign='left'),sort=["Photograph","Print","Drawing","Painting","Sculpture"])
).transform_filter(
    alt.FieldOneOfPredicate(field='Classification',oneOf=["Photograph","Print","Drawing","Painting","Sculpture"])
).configure_legend(
   orient='left',
     titleFontSize=12,
     labelFontSize=12
).configure_axis(
    titleFontSize=14,
    labelFontSize=12,
    grid=False,
    ticks=False
).configure_title(
    fontSize=18
).configure_facet(
    spacing=4
)
```

```
chart2=alt.Chart(gender_year_df,width=600).mark_circle(opacity=0).encode(
    x = alt.X("DateAcquired", 
        scale=alt.Scale(domain=[1940, 2020]),axis=alt.Axis(grid=False,title=None)),
    y = alt.Y("Title",
              scale=alt.Scale(type="log"),axis=alt.Axis(grid=False,title=None)),
    color = alt.Color("Classification")
  ).transform_filter(
    alt.FieldOneOfPredicate(field='Classification',oneOf=["Photograph","Print","Drawing","Painting","Sculpture"])
  )
#chart2
chart3=chart2+chart2.transform_regression("DateAcquired","Title",groupby=["Classification"],method="poly"
).mark_line(size=2
).encode(tooltip="Title:Q"
).interactive(
)
chart3
```

```
alt.renderers.set_embed_options(theme="dark")
alt.hconcat(chart1,chart3
).resolve_scale(color="independent"
).configure_facet(
     spacing=4
).configure_legend(
      orient='right',
      titleFontSize=12,
      labelFontSize=12
).configure_axis(
     titleFontSize=14,
     labelFontSize=12,
     ticks=False
).configure(
    #background='#1B1B1B'
).configure_view(
    stroke=None
)
```
